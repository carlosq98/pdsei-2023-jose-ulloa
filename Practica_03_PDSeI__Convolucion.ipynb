* **Importamos el audio y creamos el kernel**

````
load=files.upload()
````

````
display.Audio(load['audio2.wav'])
````

````
audio_signal, samplerate = soundfile.read('audio2.wav')
````

````
kernel = np.exp(-np.linspace(-2,2,31)**2)
kernel = kernel/sum(kernel)
N=len(audio_signal)
````

* **Mostrando señales creadas e importadas**

````
plt.figure(figsize=(15,6))

plt.subplot(211)
plt.plot(kernel,'r',linewidth=2)
plt.xlim([0,40])
plt.title('Kernel')
plt.grid()

plt.subplot(212)
plt.plot(audio_signal,'b',linewidth=2)
plt.xlim([0,N])
plt.title('Kernel')
plt.grid()

plt.show()
````

 * **Cálculo y gráfica de la convolución**

````
resultado = np.convolve (audio_signal,kernel,'same')



plt.figure(figsize=(15,3))

plt.plot(resultado,'y',linewidth=2)
plt.xlim([0,N])
plt.title('Resultados de la Convolución')
plt.grid()

plt.show()
soundfile.write('resultado.wav', resultado, samplerate)
display.Audio('resultado.wav')
````

**Comparamos**

````
plt.figure(figsize=(10, 6))
plt.plot(audio_signal, label='Audio Original')
plt.plot(resultado, label='Audio Filtrado')
plt.legend()
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.title('Audio Original vs. Audio Filtrado')
plt.show()
````

A continuación, se presentan algunas consideraciones generales:

#Correlación vs. Convolución:
Como se mencionó anteriormente, la principal diferencia entre la correlación y la convolución radica en la dirección del kernel. En la correlación, el kernel se aplica a la señal de entrada sin invertirse, mientras que en la convolución, el kernel se invierte antes de la operación.

#Efecto en el Audio:
La correlación en este contexto podría utilizarse para buscar patrones o similitudes específicas en la señal de audio. Si el kernel está diseñado para coincidir con ciertos patrones de frecuencia o características de audio, la correlación podría destacar estos patrones en la señal de salida.
